```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
#my exploration NO.1 : Line Graph on 2 plots

#data
x=np.array([0,1,2,3,4,5,6])
y1=np.array([0,0.1,0.2,0.3,0.5,0.8,1.0])
y2=np.array([1.0,2.1,3.5,6.5,7.2,5.9,6.3])

#set font dictionary to improve the labeling
font1 = {'family':'DejaVu Sans','color':'orange','size':25}
font2 = {'family':'serif','color':'darkred','size':20}

#plot line chart and improve the design of the chart
plt.plot(x,y1,marker = 'd', markersize =10, markeredgecolor = 'y', color ='hotpink', linestyle = '-.')
plt.plot(x,y2,marker = 'o', markersize =10, markeredgecolor = 'b',color = 'g', linestyle = ':')

#apply font styles to labels and title
plt.title("Plot of Data X and Y", fontdict = font1)
plt.xlabel("x-axis", fontdict = font2)
plt.ylabel("y-axis", fontdict = font2)

plt.legend(['x-axis','y-axis']) #add legend
plt.grid(color = 'g', linestyle = ':', linewidth = 0.5) #add grid lines
plt.show()
```

```python
#my exploration NO.4 : Pie Chart

#data
Workers=[40,35,100,70]
Education=['Primary','JHS','SHS','Tertiary']
Explode = [0.3,0,0,0] #explode one of the wedges in pie chart
Colors = ['limegreen','peru','lightpink','plum'] #set new colours

#set font dictionary to improve labels
font1 = {'family':'Dejavu Sans','color':'k','size':20}

#Adjust size of chart
plt.figure(figsize=(7,7))

#plot pie chart and indicate the labels and percentages while exploding one of the wedges
plt.pie(Workers, labels = Education, autopct='%1.2f%%', explode = Explode, colors = Colors, shadow = True) #display percentages at 2 decimal places

plt.legend(bbox_to_anchor=(1.2, 1),loc ='upper left',) #locate the legend at the upper right corner of the chart

plt.title('LEVELS OF EDUCATION AT CENTRAL REGION IN 2021', fontdict = font1) #add title
plt.show()
```

## Insights drawn from Pie Chart above

- Workers from the SHS level comprised of almost double the number of Tertiary students within this period; this is indicative of a large volume of unskilled labour at this education level
- The low number of JHS and Primary students is indicative of a lack in certain skills within this level group which tends to make them less suitable for certain roles
- The best stage to hold most forms of campaign that is relevant to children and young adults in this region will be at the SHS level since one is more likely to leverage on that audience who have been exposed to a working environment in addition to their academic one and may heighten their level of curiosity and zeal to take on new projects and assignments
- The barriers to entry at the Tertiary levels need to be identified and managed correctly so as to promote a balance in the transition phases and increase the chances for more of the youth to receive basic education and knowledge that would eventually equip them to become suitably employable and/or entrpreneurial as much as possible

```python
#my exploration No. 6 : Scatter Plot

#data sets
xt=np.array([0,0.1,0.2,0.3,0.5,0.8,1.0])
yt=np.array([1.0,2.1,3.5,6.5,7.2,5.9,6.3])
zt= xt*yt

#Adjust size of chart
plt.figure(figsize=(7,5))

#set title font to size 20
plt.rc('axes', titlesize=20) 


#plot scatter graph
plt.scatter(xt,yt, marker = 's') #apply square markers
plt.scatter(xt,zt, marker = '<') #apply left triangle markers 
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(['yt points','zt points'], loc = 'upper left') #position lengend at upper left corner
plt.grid() #set grid lines

plt.title('Scatter Plot')
plt.show()
```

```python
#my exploration No.9 : Grouped Bar Chart

#data
#x-axis
countries=['Ghana','Portugal','South Korea','Uruguay','Germany']

#y-axis
gold=[46,27,26,19,17]
silver=[37,23,18,18,10]
bronze=[38,30,25,18,24]

#Adjust size of chart
plt.figure(figsize=(7,7))

#bar chart properties
x = np.arange(len(countries))
width = 0.2

#draw grouped bar chart
a = plt.bar(x-width,gold,width,color='gold', label = 'Gold')
b = plt.bar(x,silver,width,color='silver', label ='Silver')
c = plt.bar(x+width,bronze, width,color='brown', label = 'Bronze')

#plt.xlabel('Countries')
plt.ylabel('Number of Trophies Won')
plt.xticks(x, countries) #gets location for the labels on x-axis
plt.legend()


plt.title('Trophy Number and Type Won by 5 Countries in Olympics 2016')
plt.show()
```

## Insights drawn from Grouped Bar Chart above

- The majority of the trophies won by the select 6 countries were Gold and Bronze.
- The 2016 Olympics was held in Rio de Janeiro, Brazil. The weather in this country is largely tropical.
- The countries that won more gold trophies than the otheres were Ghana, Suth Korea and Uruguay. Other than South Korea, where the weather is temperate, Ghana and Uruguay appeared to have had the benefit of the climate type on their side spuring them onto winning more gold trophies than the rest. These 2 countries are tropical in climate, thus possibly fostering the athletes to be comfortable enough to perform their best in somewhat 'familiar' terrain.
- It may be safe to say the location for the olympics may have an impact in the outcome of the performance by the various participating countries.

## More insights
- Ghana participated in more games than the oter countries, hence more trophies
